## Using HF pipeline() 

```python
from transformers import pipeline, infer_device
device = infer_device()
```

```python
# Loding model and importing gradio
import gradio as gr
generator = pipeline(
    "text-generation",
    model="codelion/gpt-2-70m",
    device=device
)
```

## Function for gradio

```python
# function for gradio
def generate_response(user_input):
  output = generator(
      user_input,
      max_length=80,
      num_return_sequences=1,
      truncation=True
  )

  return output[0]["generated_text"]


with gr.Blocks() as demo:
  gr.Markdown("## ðŸ”¥ Simple GPT-2 Chat Interface")

  with gr.Row():
        user_input = gr.Textbox(
            label="Your Message",
            placeholder="Type something...",
            lines=2
        )
  generate_btn = gr.Button("Generate")
    
  output_box = gr.Textbox(
        label="Model Response",
        lines=5
    )
  
  

  generate_btn.click(
      fn=generate_response,
      inputs=user_input,
      outputs=output_box
    )

demo.launch(share=True)
```
